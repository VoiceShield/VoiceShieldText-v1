# Removing Punctuations, Special Characters and Digits

```python
import unicodedata

def remove_punctuation_unicode(text):

    text = str(text)

    cleaned_text = ""
    removed_characters = []

    for ch in text:
        category = unicodedata.category(ch)
        if category.startswith("P"):
            removed_characters.append((ch, category))
        else:
            cleaned_text += ch
    if removed_characters:
      print(f"Removed Characters categorized as punctuation:")
    for ch, category in removed_characters:
        print(f"Character: {ch}, Category: {category}")

    return cleaned_text

# Example usage
text1 = "አልተሻልክም።"
clean_text1 = remove_punctuation_unicode(text1)
print(f"Cleaned Text 1: {clean_text1}")

text2 = 123.45  # Example with float
clean_text2 = remove_punctuation_unicode(text2)
print(f"Cleaned Text 2: {clean_text2}")
```

```python
# special symbol removal
df_final['text']=df_final['text'].str.replace('[^\w\s]','',regex=True)
# removing digits from the dataset
df_final['text']=df_final['text'].str.replace('\d+','',regex=True)
```

```python
# apply normalization and special characters removal
df_final['text']=df_final['text'].apply(lambda x: remove_punctuation_unicode(x))

df_final.head()
```

```python
def remove_punctuation_mapping(text):
    punctuation_mapping = {
        "፡": "",
        "።": "",
        "፣": "",
        "፤": "",
        "፥": "",
        "፦": "",
        "፧": "",
        "፨": "",
        "!": "",
        "?": "",

    }
    return "".join([punctuation_mapping.get(ch, ch) for ch in text])
```

```python
df_final['text']=df_final['text'].apply(lambda x: remove_punctuation_mapping(x))
```

# Removing stop words

```python
import re
path = '/content/drive/MyDrive/stopwords-am.txt'
words_to_remove = set()
with open(path, 'r', encoding='utf-8') as file:
    words_to_remove = {word.strip() for word in file.readlines()}

# Function to remove words from each sentence
def remove_words_from_sentence(sentence):
    # Tokenize the sentence by spaces or punctuation except Amharic-specific punctuation
    tokens = re.findall(r'\w+|[^\w\s]', sentence, flags=re.UNICODE)
    # Filter out tokens that are in the words_to_remove set
    filtered_tokens = [token for token in tokens if token not in words_to_remove]
    # Join the tokens back into a cleaned sentence
    cleaned_sentence = ' '.join(filtered_tokens)
    return cleaned_sentence

# Apply the function to clean each sentence in the 'text' column
df_final['cleaned_text'] = df_final['text'].apply(remove_words_from_sentence)

# Print the updated DataFrame
print(df_final)
```
